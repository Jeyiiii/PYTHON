## Setting Attributes


```python
class anime:
    
    def char (self):
        print("One Piece")
    
    def show_char (self, char):
        self.char = char
    
    def get_char (self):
        return self.char

x = anime()
x.char()
x.show_char("Luffy")
print(x.get_char())

```

## Using the __init__ function





```python
class anime:
    
    def __init__ (op, title,author, protag, goal,):
        op.title = title
        op.author = author
        op.protag = protag
        op.goal = goal
        
    def show(op):
        print('Title:', op.title)
        print('Author:', op.author)
        print('Protagonist:', op.protag)
        print('Goal:', op.goal)
        
    def intro(op):
        print (op.title, 'is a Japanese Manga Series written and illustrated by', op.author , '. It follows the adventure of', op.protag , 'and his crew to find the One Piece Treasure and achieve his dream of becoming', op.goal,'.')

x = anime ('One Piece', 'Eiichiro Oda', 'Monkey D. Luffy', 'King of the Pirates') 
x.show()
x.intro()
        
        
        
```

## Use of classes, objects, and methods


```python
class onepiece:
    
    def __init__ (pirates, captain1, cook, captain2, swordsman, number2):
        pirates.captain1 = captain1
        pirates.cook = cook
        pirates.captain2 = captain2
        pirates.swordsman = swordsman
        pirates.number2 = captain2
        
        print("FAN FAVORITE CHARACTERS FROM ONE PIECE:")
        
    def show(pirates): 
        print('Heart Pirates Captain:', pirates.captain1)
        print("Strawhat's Cook:", pirates.cook)
        print('Strawhat Pirates Captain:', pirates.captain2)
        print("Strawhat's Swordsman:", pirates.swordsman)
        print('Revolutionary Army Chief of Staff:', pirates.number2)
        
x = onepiece ('Trafalgar D. Water Law', 'Vinsmoke Sanji', 'Monkey D. Luffy', 'Roronoa Zoro', 'Sabo' )
x.show()

class demonslayer:
    
    def __init__(hashira, mist, sound, wind, serpent, love, flame, insect, stone, water):
        hashira.mist = mist
        hashira.sound = sound
        hashira.wind = wind
        hashira.serpent = serpent
        hashira.love = love
        hashira.flame = flame
        hashira.insect = insect
        hashira.stone= stone
        hashira.water = water
        
        print("")
        print("HASHIRAS FROM KIMETSU NO YAIBA:")
        
    def show(hashira):
        print('Mist Hashira:', hashira.mist)
        print('Sound Hashira:', hashira.sound)
        print('Wind Hashira:', hashira.wind)
        print('Serpent Hashira:', hashira.serpent)
        print('Love Hashira:', hashira.love)
        print('Flame Hashira:', hashira.flame)
        print('Insect Hashira:', hashira.insect)
        print('Stone Hashira:', hashira.stone)
        print('Water Hashira:', hashira.water)
        
x = demonslayer ('Muichiro Tokito', 'Tengen Uzui', 'Sanemi Shinazugawa', 'Obanai Iguro', 'Mitsuri Kanjori', 'Kyojuro Rengoku', 'Shinobu Kocho', 'Gyomei Himejima', 'Giyu Tomioka' )
x.show()   

class dragonball:
    
    def __init__(saiyan, s1, s2, s3, s4, s5, s6, s7, s8,):
        saiyan.s1 = s1
        saiyan.s2 = s2
        saiyan.s3 = s3
        saiyan.s4 = s4
        saiyan.s5 = s5
        saiyan.s6 = s6
        saiyan.s7 = s7
        saiyan.s8 = s8
        
        print("")
        print("SAIYANS FROM DRAGONBALL:")
        
    def show(saiyan):
        print(saiyan.s1)
        print(saiyan.s2)
        print(saiyan.s3)
        print(saiyan.s4)
        print(saiyan.s5)
        print(saiyan.s6)
        
        print("FUSED SAIYANS:")
    def fusion(saiyan):
        print(saiyan.s7)
        print(saiyan.s8)
        
x = dragonball ('Vegeta', 'Goku', 'Gohan', 'Trunks', 'Young Gohan', 'Young Trunks', 'Gotenks', 'Vegito',)
x.show()
x.fusion()

class jojo:
    
    def __init__(joestar, j1, j2, j3, j4, j5, j6,):
        joestar.j1 = j1
        joestar.j2 = j2
        joestar.j3 = j3
        joestar.j4 = j4
        joestar.j5 = j5
        joestar.j6 = j6
        
        print("")
        print("PROTAGONISTS FROM JoJo's BIZZARE ADVENTURES:")
        
    def show(joestar):
        print(joestar.j1)
        print(joestar.j2)
        print(joestar.j3)
        print(joestar.j4)
        print(joestar.j5)
        print(joestar.j6)
         
x = jojo ('Jotaro Kujo', 'Jonathan Joestar', 'Dio Brando', 'Joseph Joestar', 'Rohan Kishibe', 'Erina Pendleton')
x.show()

class naruto:
    
    def __init__(hokage, h1, h2, h3, h4, h5, h6, h7):
        hokage.h1 = h1
        hokage.h2 = h2
        hokage.h3 = h3
        hokage.h4 = h4
        hokage.h5 = h5
        hokage.h6 = h6
        hokage.h7 = h7
        
        print("")
        print("HOKAGES FROM NARUTO SHIPPUDEN:")
        
    def show(hokage):
        print('First Hokage:', hokage.h1)
        print('Second Hokage:', hokage.h2)
        print('Third Hokage:', hokage.h3)
        print('Fourth Hokage:', hokage.h4)
        print('Fifth Hokage:', hokage.h5)
        print('Sixth Hokage:', hokage.h6)
        print('Seventh Hokage:', hokage.h7)
         
x = naruto ('Hashirama Senju', 'Tobirama Senju', 'Hiruzen Sarutobi', 'Minato Namikaze', 'Tsunade Senju', 'Kakashi Hatake', 'Naruto Uzumaki')
x.show()


```
